````python
def simulasi_pot_bunga(n,m):
  # Kompleksitas O(n^2)
  # membuat list pot_bunga yang isinya adalah string 0 yang menandakan bahwa seluruh tanaman belum tersiram, # posisi: sebagai iterator
  pot_bunga, posisi = ['0']*n, 0 
  # loop dari 0 - n. batas atasnya adalah n karena maksimal seluruh bunga tersiram adalah sebanyak n kali.
  for bunga in range(n):
    # agar bunga yang sudah disiram tidak disiram lagi
    if pot_bunga[posisi] == '1':
      break
    # menyiram bunga pada index posisi
    pot_bunga[posisi] = '1'
    # menampilkan simulasi penyiraman bunga
    print(''.join(pot_bunga),f'| bunga ke-{posisi} disiram')
    # memperbaharui posisi
    posisi = (posisi + m) % n 
    # mengembalikan jumlah 0 yang terdapat pada list   pot_bunga
  return pot_bunga.count('0')

# Menampilkan Simulasi Penyiraman untuk 8 Pot Bunga dengan M dari 1 - 7 dan untuk mencari pola agar algoritma bisa dioptimisasi.
print('SIMULASI PENYIRAMAN\nN = 8\n')
for i in range(1,8):
  print(f'Jika M = {i} maka jumlah tanaman yang tidak tersiram adalah {simulasi_pot_bunga(8,i)}\n')
````

Tuliskan program untuk menyelesaikan masalah Pasangan Titik Terdekat dengan strategi Exhaustive Search. 

Permasalahan

Ingat kembali bahwa dalam bidang Cartesius, jarak antar dua buah titik $P_1=(x_1,y_1)$ dan $P_2=(x_2,y_2)$ didefinisikan oleh jarak Euclid sebagai berikut:

$$
d(P_1,P_2) = \sqrt{(x_1-x_2)^2+(y_1-y_2)^2}.
$$

Dalam masalah ini, Anda akan diberikan input berupa dua buah vektor ```X``` dan ```Y``` yang masing-masing berukuran ```n``` dan berisi bilangan real. Dua buah vektor tersebut menyatakan ```n``` buah titik pada bidang Cartesius, yaitu ``` (x[0], y[0]), (x[1], y[1]), ... , (x[n-1], y[n-1])```. Anda diminta untuk menghasilkan output berupa **semua** pasangan titik yang memiliki jarak terdekat dari antara semua pasangan titik lainnya.

```
Input  : Two lists x and y of n (n≥2) points p1(x1,y1),...,pn(xn,yn)
Output : The distance between the closest pair of points and all pairs of points attaining that distance.
```


````python
from numpy import inf


def find_distance(x1, y1, x2, y2):
  # fungsi untuk mencari jarak antar dua titik
  return ((x1-x2)**2 + (y1-y2)**2)**0.5


def closest_pair(x, y):
  # membuat jarak terkecil menjadi sangat besar agar nanti saat iterasi ditemukan
  # jarak yang lebih kecil variabel min_d akan diupdate
  min_d = inf
  koordinat = []
  for i in range(len(x)):
    for j in range(i+1, len(x)):
      distance = find_distance(x[i], y[i], x[j], y[j])
      # mengupdate min_d jika distance < min_d
      # dan list koordinat menjadi index i dan j
      if distance < min_d:
        min_d = distance
        koordinat = [[i, j]]
      elif distance == min_d:  # jika ada lebih dari satu titik yang memiliki jarak terkecil
        # simpan jarak terkecil di list koordinat
        koordinat.append([i, j])

  # Output
  print(round(min_d,3))
  for i, j in koordinat:
    print(i, j)


def main():
  print('Input:')
  sb_x = list(map(float, input().split(',')))
  sb_y = list(map(float, input().split(',')))
  print('\nOutput:')
  closest_pair(sb_x, sb_y)


if __name__ == '__main__':
  main()
````

Kompleksitas
* Kompleksitas waktu $=\mathcal{O}(n^2)$
* Kompleksitas memori $=\mathcal{O}(n)$

````python
# ukuran_matrix = int(input('Ukuran Matrix (n x n): '))
# treasure_matrix = [list(map(int,input().split(' '))) for i in range(ukuran_matrix)]
treasure_matrix = [[1,0,1000],
       [5,0,0],
       [5,5,5]
       ]
ukuran_matrix = 3
````

STRATEGI DFS WITH RECURSION

````python
def dfs_treasure_hunt(row,col,matrix,ukuran_matrix,k = 2*(ukuran_matrix - 1) ,path = ''):
  # Base Case: Saat posisi sudah ada di pojok kanan bawah
  if (row == ukuran_matrix - 1 and col == ukuran_matrix - 1) or k == 0:
    return matrix[row][col],path  

  # Depth First Search
  candidate_solution = [[-1],[-1]] 
  # index 0: simpan hasil rekursi ke bawah
  # index 1: simpan hasil rekursi ke kanan

  # jika baris berikutnya < dari ukuran matriks (memungkinkan kebawah)
  # simpan hasil rekursi kebawah ke candidate solution
  # kalau tidak bisa, maka candidate solution untuk kebawah adalah 0
  next_row = row + 1
  if next_row < ukuran_matrix:
    candidate_solution[0] = dfs_treasure_hunt(next_row,col,matrix,ukuran_matrix,k-1,path+'B')
  
  # jika kolom berikutnya < dari ukuran matriks (memunkinkan kekanan)
  # simpan hasil rekursi kekanan ke candidate solution
  # kalau tidak bisa, maka candidate solution untuk kebawah adalah 0
  next_col = col + 1
  if next_col < ukuran_matrix:
    candidate_solution[1] = dfs_treasure_hunt(row, next_col, matrix,ukuran_matrix,k-1,path+'K')
  
  # jika hasil rekursi ke bawah lebih besar dari hasil rekursi ke kanan
  if candidate_solution[0][0] > candidate_solution[1][0]:
    # tambahkan candidate_solution yang menyimpan hasil kebawah dengan harta yang ada di posisi saat itu dan rutenya
    return candidate_solution[0][0] + matrix[row][col],candidate_solution[0][1]
  # jika hasil rekursi ke kanan lebih besar dari hasil rekursi ke bawah
  else:
    return candidate_solution[1][0] + matrix[row][col],candidate_solution[1][1]

harta_karun = dfs_treasure_hunt(0,0,treasure_matrix,ukuran_matrix)
print(f'Harta terbanyak yang didapatkan adalah sebanyak {harta_karun[0]} dengan rute {harta_karun[1]}.')
````

STRATEGI DFS DENGAN STACK

````python
def dfs_treasure_hunt_stack(matrix, ukuran_matrix, k=2*(ukuran_matrix-1)):
    stack = []
    max_gold = 0
    best_rute = ''
    # Inisialisasi
    # x = row, y = col
    # Stack isinya posisi x,y,path,isi posisinya itu,sisa langkah 
    stack.append([0, 0, '', matrix[0][0], k])
    while stack:
      # Telusuri
        x, y, rute, val, k = stack.pop()

        # Check apakah dia sudah di kanan bawah atau tidak
        if (x == ukuran_matrix-1 and y == ukuran_matrix-1) or k == 0:
            if val > max_gold:
                max_gold = val
                best_rute = rute
        else:
          # Kalau dia masih bisa kebawah 
            if x + 1 < ukuran_matrix:
                stack.append([x+1, y, rute+"B", val+matrix[x+1][y], k-1])
          # Kalau dia masih bisa kekanan
            if y + 1 < ukuran_matrix:
                stack.append([x, y+1, rute+"K", val+matrix[x][y+1], k-1])

    print(f'Harta terbanyak yang didapatkan adalah sebanyak {max_gold} dengan rute {best_rute}.')
dfs_treasure_hunt_stack(treasure_matrix,ukuran_matrix,6)
````

STRATEGI BFS DENGAN QUEUE

````python
def BFS_treasure_hunt_queue(matrix, ukuran_matrix, k=2*(ukuran_matrix-1)):
    queue = []
    max_gold = 0
    best_rute = ''
    # Inisialisasi
    queue.append([0, 0, '', matrix[0][0], k])
    while queue:
        x, y, rute, val, k = queue.pop(0)
        if (x == ukuran_matrix-1 and y == ukuran_matrix-1) or k == 0:
            if val > max_gold:
                max_gold = val
                best_rute = rute
        else:
            if x + 1 < ukuran_matrix:
                queue.append([x+1, y, rute+'B', val+  matrix[x+1][y], k-1])
            if y + 1 < ukuran_matrix:
                queue.append([x, y+1, rute+'K', val + matrix[x][y+1], k-1])
    print(f'Harta terbanyak yang didapatkan adalah sebanyak {max_gold} dengan rute {best_rute}.')
BFS_treasure_hunt_queue(treasure_matrix,ukuran_matrix,6)
````

````python
def naik_lift(c,arr,awal=0,ans=[],count=0,berat=0):
  # BASE CASE --> Kalau beratnya sudah melebihi dari daya tampung artinya tidak ada yang bisa naik lagi
  if berat > c:
    return 0, []
  
  if len(arr) == 0:
    return count, ans

  max_count = count
  max_ans = ans

  for i in range(len(arr)):
    temp_count, temp_answer = naik_lift(c,arr[i+1:],awal+i+1,ans+[awal+i],count+1,berat+arr[i])
    if temp_count > max_count:
      max_count = temp_count
      max_ans = temp_answer
  return max_count, max_ans

if __name__ == '__main__':
  # c = int(input('Daya Tampung Lift: '))
  c = 1000
  # berat_mahasiswa = list(map(int,input().split(' ')))
  berat_mahasiswa = [60 ,87 ,53 ,90 ,88 ,62 ,48 ,65 ,72 ,85,50,51,78,79,89,91,92,48,49,43]
  ans = naik_lift(c,berat_mahasiswa)
  berat = [berat_mahasiswa[i] for i in ans[1]]
  total_berat = sum(berat)
  berat = list(map(str,berat))
  berat = ' '.join(berat)
  print(f'Daya tampung Lift adalah {c} kg.')
  print(f'Total Mahasiswa yang dapat naik lift adalah {ans[0]}.')
  print(f'Mahasiswa yang naik lift adalah mahasiswa dengan berat {berat} dengan total berat {total_berat} kg.')
````

````python
def GUI():
  print('-'*80,'\n\n')

def fake_coin(fake, end, start=0, timbang=1):
    num_coins = end - start
    cut_three = num_coins//3
    # BASE CASE
    if num_coins == 1:
        print('-'*80)
        print(f'Ketemu! Koin palsu adalah koin {start}.')
        print(f'\t Total penimbangan yang dilakukan: {timbang - 1}.')
        return

    if num_coins == 2:
        print(f'Koin {start} diletakkan di sebelah kiri timbangan.')
        print(f'Koin {end-1} diletakkan di sebelah kanan timbangan.')

        if fake == start:
            print(f'Koin palsu berada di tangan sebelah kiri timbangan!')
            print(f'\t Hasil Penimbangan {timbang} : Timbangan sebelah kiri lebih ringan.')
            GUI()
            timbang += 1
            fake_coin(fake, (start+1), start, timbang)
            return

        else:
            print(f'Koin palsu berada di tangan sebelah kanan timbangan!')
            print(f'\t Hasil Penimbangan {timbang} : Timbangan sebelah kanan lebih ringan.')
            GUI()
            timbang += 1
            fake_coin(fake, end, (start+1), timbang)
            return

    print('-'*80)
    print(f'Sekarang kita tinjau koin {start} hingga {end-1}.\n')
    print(
        f'Letakan koin {start} hingga {start+cut_three-1} di sebelah kiri tangan timbangan.')
    print(
        f'Letakan koin {start+cut_three} hingga {start+cut_three+cut_three-1} di sebelah kanan tangan timbangan.\n')
    print(f'Sisihkan koin {start+cut_three+cut_three} hingga {end-1}.')

    # DECREASE AND CONQUER
    if start <= fake < (start+cut_three):  # Lengan Kiri Lebih Ringan
        print('Koin palsu berada di tangan sebelah kiri timbangan!')
        print(
            f'\t Hasil Penimbangan {timbang} : Timbangan sebelah kiri lebih ringan.')
        GUI()
        timbang += 1
        fake_coin(fake, (start+cut_three), start, timbang)

    elif (start+cut_three) <= fake < start+cut_three+cut_three:  # Lengan Kanan Lebih Ringan
        print('Koin palsu berada di tangan sebelah kanan timbangan!')
        print(
            f'\t Hasil Penimbangan {timbang} : Timbangan sebelah kanan lebih ringan.')
        GUI()
        timbang += 1
        fake_coin(fake, start+cut_three+cut_three, start+cut_three, timbang)

    else:  # Balance
        print('Koin palsu berada pada kelompok ketiga!')
        print(f'\t Hasil Penimbangan {timbang} : Seimbang!')
        GUI()
        timbang += 1
        fake_coin(fake, end, start+cut_three+cut_three, timbang)


if __name__ == '__main__':
    jumlah_koin = int(input('Inputkan banyak koin yang akan ditimbang: '))
    koin_palsu = int(
        input(f'Koin ke berapa yang palsu? (koin 0 - koin {jumlah_koin-1}): '))
    GUI()
    fake_coin(koin_palsu, jumlah_koin)
````

Implementasikan Algoritma Quick Sort dengan efisiensi waktu rata-rata $\mathcal{O}(n \log n)$ dan efisiensi ruang $\mathcal{O}(n \log n)$.

````python
def quick_sort(start, end, arr):
  # BASE CASE
    if start < end:
        partition_pos = partition(start, end, arr)
        quick_sort(start, partition_pos, arr) # Quicksort Left Side
        quick_sort(partition_pos + 1, end, arr) # Quicksort Right Side

def partition(start, end, arr):
    pivot = arr[start]
    i = start + 1
    j = end - 1

    while i < j:
        while arr[i] <= pivot and i < end - 1:
            i += 1

        while arr[j] > pivot and j > start:
            j -= 1

        if i < j:
            arr[i], arr[j] = arr[j], arr[i]

    if arr[j] < pivot:
        arr[j], arr[start] = arr[start], arr[j]

    return j

if __name__ == '__main__':
    from random import randint
    for i in range(10):
        l = []
        for j in range(10):
            num = randint(-10,10)
            l.append(num)
        print(f'Unsorted = {l}')
        p_sort = sorted(l)
        print(f'Python Sort = {p_sort}')
        quick_sort(0,len(l),l)
        print(f'QuickSort = {l}\n')
        assert l == p_sort
````

Implementasikan Algoritma Strassen Matrix Multiplication dengan efisiensi waktu $\mathcal{O}(n^{\log_2 7})$ dan efisiensi ruang $\mathcal{O}(n^2 \log n)$

````python
M1 = [[3,6,2,10,5],[31,1,5,3,5],[3,3,3,3,5],[3,6,2,10,5],[31,1,5,3,5]]
M2 = [[3,7,2,2,2],[5,1,2,3,4],[6,2,3,3,2],[4,2,1,4,2],[3,6,2,10,5]]

# Untuk Memastikan Bahwa Matrix itu 2^n x 2^n
def setup(M1,M2,n,r,c):
  # r = [r1,r2]
  # c = [c1,c2]
  # Tambah 0 di setiap ujung baris nya agar baris nya jadi ukuran bilangan kuadrat kelipatan 2 --> Matrix 1
  for i in range(r[0]):
    M1[i]+=[0]*(n-c[0])

  # Tambah 0 di setiap ujung baris nya agar baris nya jadi ukuran bilangan kuadrat kelipatan 2 --> Matrix 2
  for i in range(r[1]):
    M2[i]+=[0]*(n-c[1])

  # Tambah 0 di setiap ujung kolom nya agar kolom nya jadi ukuran bilangan kuadrat kelipatan 2 --> Matrix 1
  for i in range(n-r[0]):
    M1.append([0]*n)

  # Tambah 0 di setiap ujung kolom nya agar kolom nya jadi ukuran bilangan kuadrat kelipatan 2 --> Matrix 2
  for i in range(n-r[1]):
    M2.append([0]*n)

# Cari Pangkat Agar Matrix Nya Bisa n x n
def find_n(M1,M2):
  # Mencari Ukuran Matrix Pertama dan Kedua
  r1,c1 = len(M1),len(M1[0])
  r2,c2 = len(M2),len(M2[0])

  # Kolom M1 != Baris M2
  if c1!=r2:
    raise TypeError("Matriks tidak bisa dikalikan")


  p = max(r1,r2,c1,c2)
  n = 1

# Buat Cari Bilangan Pangkat 2 Terdekat yang Lebih Besar Dari P
  while n<p:
    n*=2
  return n,n>p,(r1,r2),(c1,c2)

def bagi4(M,n):
  a = [M[_][:n//2] for _ in range(n//2)]
  b = [M[_][n//2:] for _ in range(n//2)]
  c = [M[_][:n//2] for _ in range(n//2,n)]
  d = [M[_][n//2:] for _ in range(n//2,n)]
  return a,b,c,d

def gabung(a,b,c,d,n,first,R,C):
  # List Buat Simpan Hasil Akhir
  M = []
  for i in range(n//2):
    M.append(a[i]+b[i])
  for i in range(n//2):
    M.append(c[i]+d[i])
  # Kalau dia first kita buang 0 nya. Biar hasil akhirnya udh g ada 0 0 nya
  if first:
    M = [arr[:C[1]] for arr in M[:R[0]]]  
  return M

# Fungsi Nambahin Matrix
def tambah(a,b):
  n = len(a)
  a_b = [[0]*n for i in range(n)]
  for i in range(n):
    for j in range(n):
      a_b[i][j] = a[i][j] + b[i][j]
  return a_b

# Fungsi Kurangin Matrix
def kurang(a,b):
  n = len(a)
  a_b = [[0]*n for i in range(n)]
  for i in range(n):
    for j in range(n):
      a_b[i][j] = a[i][j] - b[i][j]
  return a_b

# Fungsi Perkalian Matrix Pake Metode Strasen
def kali(M1,M2,first=False):
  # First itu buat nandadin kalau dia rekursi paling awal atau bukan
  # BASE CASE --> Pecah Matrix sampai ukuran 1x1
  if len(M1) == 1 == len(M2):
    return [[M1[0][0]*M2[0][0]]]

  # Deteksi Apakah M1 == M2.
  # Bug Prevention. Python itu anggep kalau M1 sama persis kayak M2
  # Pas Setup nya bakal error karena kalau kita nambah kolom di M1
  # si python anggep M2 itu M1 makanya is M2 juga ketambahan kolom
  # Maka dari itu kalau dia sama kita buat list baru tapi isinya sama
  if id(M1[0]) == id(M2[0]):
    M2 = [_[:] for _ in M2]

  # R_ = [R1,R2]
  # C_ = [C1.C2]
  n,need_setup,R_,C_ = find_n(M1,M2)
  # Kalau Ukuran nya dia bukan bilangan kuadrat 2 maka kita butuh setup (tambahin 0)
  if need_setup: 
    setup(M1,M2,n,R_,C_)

  a1,b1,c1,d1 = bagi4(M1,n)
  a2,b2,c2,d2 = bagi4(M2,n)

  # STRASSEN
  m1 = kali(tambah(a1,d1),tambah(a2,d2))
  m2 = kali(tambah(c1,d1),a2)
  m3 = kali(a1,kurang(b2,d2))
  m4 = kali(d1,kurang(c2,a2))
  m5 = kali(tambah(a1,b1),d2)
  m6 = kali(kurang(c1,a1),tambah(a2,b2))
  m7 = kali(kurang(b1,d1),tambah(c2,d2))

  A = tambah(kurang(tambah(m1,m4),m5),m7)
  B = tambah(m3,m5)
  C = tambah(m2,m4)
  D = tambah(kurang(tambah(m1,m3),m2),m6)

  return gabung(A,B,C,D,n,first,R_,C_)

for _ in kali(M1,M2,True):
  print(_)
````

Metode Strasen:
Pertama , karena algoritma ini memakai rekursi maka base case nya dalah ketika ukura matrix nya itu 1x1. Kalau matrix itu sudah 1x1 maka kita kali aja. Sebelom matrix dikalikan kita perlu check apakah matrix itu bisa dikalikan atau engga. Syarat matrix bisa dikalikan atau engga itu adalah Kolom matrix pertama harus = Baris Matrix Kedua dan Ukuran matrix nya harus persegi berukuran bilangan kuadrat 2. Fungsi Find _ N itu untuk cek apakah dia bisa dikalikan atau engga. Kalau dia butuh setup ya kita setup kalau engga ya gausa berarti dia udh bener.

Langkah selanjutnya kita bagi 4 matrix nyaa. Hasil bagi disimpen ke a1,b1,c1,d1. Terus kita tinggal pake rumus si strasen. Kalo tambah kurang ya tambah kurang biasa aja kalo kali kita pake rekursi nya strasen tar simpan di A B C D terus digabung
